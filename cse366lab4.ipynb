Genetic algorithm


```python
import random

def genchrom(size):
    chrom = []
    for i in range(size):
        chrom.append(random.randint(1, size))
    return chrom


def crossOver(chrom1, chrom2, size):
    k = random.randint(2, size - 1)

    print("Crossover Point =", k)

    c1 = chrom1[:k] + chrom2[k:]
    c2 = chrom2[:k] + chrom1[k:]

    return c1, c2


def mutation(chrom):
    size = len(chrom)

    position = random.randint(0, size - 1)
    value = random.randint(1, size)

    chrom[position] = value

    return chrom


def fitness(chrom):
    size = len(chrom)

    maxpossibleattack = size * (size - 1) // 2
    attack = 0

    for i in range(size):
        for j in range(i + 1, size):

            # same row attack
            if chrom[i] == chrom[j]:
                attack += 1

    return maxpossibleattack - attack



population_size = 6
chromosome_size = 8

population = []

for i in range(population_size):
    population.append(genchrom(chromosome_size))
#algo
for generation in range(6):
    print("Generation", generation + 1)
    for chrom in population:
        print(chrom, "Fitness =", fitness(chrom))
    new_population = []

    while len(new_population) < population_size:

        # Random parent selection
        parent1 = random.choice(population)
        parent2 = random.choice(population)

        # Crossover
        child1, child2 = crossOver(
            parent1,
            parent2,
            chromosome_size
        )

        # Mutation
        child1 = mutation(child1)
        child2 = mutation(child2)

        # Add children
        new_population.append(child1)

        if len(new_population) < population_size:
            new_population.append(child2)

    population = new_population

print("\nFinal Population:")
# fitness break
for chrom in population:
    print(chrom, "Fitness =", fitness(chrom))
```
