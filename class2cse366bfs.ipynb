BFS

```python
from queue import Queue

class GridEnv:
    def __init__(self, size, goal, agentPos, visited):
        self.size = size
        self.goal = goal
        self.agentPos = agentPos
        self.visited = visited

    def is_goal(self):
        return self.goal == self.agentPos


class Agent007:
    def __init__(self, env):
        self.env = env
        self.agentX = self.env.agentPos[0]
        self.agentY = self.env.agentPos[1]

    def perceive(self, q):
        data = []

        # Up
        if (
            self.agentY - 1 > 0
            and [self.agentX, self.agentY - 1] not in self.env.visited
            and [self.agentX, self.agentY - 1] not in list(q.queue)
        ):
            data.append([self.agentX, self.agentY - 1])

        # Left
        if (
            self.agentX - 1 > 0
            and [self.agentX - 1, self.agentY] not in self.env.visited
            and [self.agentX - 1, self.agentY] not in list(q.queue)
        ):
            data.append([self.agentX - 1, self.agentY])

        # Down
        if (
            self.agentY + 1 <= self.env.size
            and [self.agentX, self.agentY + 1] not in self.env.visited
            and [self.agentX, self.agentY + 1] not in list(q.queue)
        ):
            data.append([self.agentX, self.agentY + 1])

        # Right
        if (
            self.agentX + 1 <= self.env.size
            and [self.agentX + 1, self.agentY] not in self.env.visited
            and [self.agentX + 1, self.agentY] not in list(q.queue)
        ):
            data.append([self.agentX + 1, self.agentY])

        # Up-Left
        if (
            self.agentX - 1 > 0
            and self.agentY - 1 > 0
            and [self.agentX - 1, self.agentY - 1] not in self.env.visited
            and [self.agentX - 1, self.agentY - 1] not in list(q.queue)
        ):
            data.append([self.agentX - 1, self.agentY - 1])

        # Up-Right
        if (
            self.agentX + 1 <= self.env.size
            and self.agentY - 1 > 0
            and [self.agentX + 1, self.agentY - 1] not in self.env.visited
            and [self.agentX + 1, self.agentY - 1] not in list(q.queue)
        ):
            data.append([self.agentX + 1, self.agentY - 1])

        # Down-Left
        if (
            self.agentX - 1 > 0
            and self.agentY + 1 <= self.env.size
            and [self.agentX - 1, self.agentY + 1] not in self.env.visited
            and [self.agentX - 1, self.agentY + 1] not in list(q.queue)
        ):
            data.append([self.agentX - 1, self.agentY + 1])

        # Down-Right
        if (
            self.agentX + 1 <= self.env.size
            and self.agentY + 1 <= self.env.size
            and [self.agentX + 1, self.agentY + 1] not in self.env.visited
            and [self.agentX + 1, self.agentY + 1] not in list(q.queue)
        ):
            data.append([self.agentX + 1, self.agentY + 1])

        return data


# Create Environment
g1 = GridEnv(4, [1, 1], [3, 1], [])

# Create Agent
masudRana = Agent007(g1)

# BFS Queue
q = Queue()

# Put starting position in queue
q.put(g1.agentPos)

while not q.empty():

    agent_current_state = q.get()

    if agent_current_state in g1.visited:
        continue

    g1.visited.append(agent_current_state)

    masudRana.agentX = agent_current_state[0]
    masudRana.agentY = agent_current_state[1]

    print("Masud Rana in:", agent_current_state)

    if g1.goal == agent_current_state:
        print("Goal Reached")
        break

    data = masudRana.perceive(q)

    for d in data:
        q.put(d)
```
