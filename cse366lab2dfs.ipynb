```python
class GridEnv:
    def __init__(self, size, goal, agentPos, visited):
        self.size = size
        self.goal = goal
        self.agentPos = agentPos
        self.visited = visited

    def is_goal(self):
        return self.goal == self.agentPos


class Agent007:
    def __init__(self, env):
        self.env = env
        self.agentX = self.env.agentPos[0]
        self.agentY = self.env.agentPos[1]

    def perceive(self, stack):
        data = []

        directions = [
            [0, -1],   # Up
            [-1, 0],   # Left
            [0, 1],    # Down
            [1, 0],    # Right
            [-1, -1],  # Up-Left
            [1, -1],   # Up-Right
            [-1, 1],   # Down-Left
            [1, 1]     # Down-Right
        ]

        for dx, dy in directions:
            nx = self.agentX + dx
            ny = self.agentY + dy

            if (
                1 <= nx <= self.env.size
                and 1 <= ny <= self.env.size
                and [nx, ny] not in self.env.visited
                and [nx, ny] not in stack
            ):
                data.append([nx, ny])

        return data


# Create Environment
g1 = GridEnv(4, [1, 1], [3, 1], [])

# Create Agent
masudRana = Agent007(g1)

# DFS Stack
stack = []
stack.append(g1.agentPos)

while stack:

    agent_current_state = stack.pop()  # LIFO

    if agent_current_state in g1.visited:
        continue

    g1.visited.append(agent_current_state)

    masudRana.agentX = agent_current_state[0]
    masudRana.agentY = agent_current_state[1]

    print("Masud Rana in:", agent_current_state)

    if g1.goal == agent_current_state:
        print("Goal Reached")
        break

    data = masudRana.perceive(stack)

    for d in data:
        stack.append(d)
```
